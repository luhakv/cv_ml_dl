### 2. Классические задачи NLP

→ Для начала разберёмся с тем, какие задачи решает NLP-инженер.

Вообще задачи в NLP можно разделить на две большие группы:

- Работа  со звуком (когда мы работаем с записанной речью).
- Работа с текстом.

Но для работы со звуком мы просто можем перевести речь в текст методами Speech2Text и затем применить методы обработки текстов. И в курсе в основном мы будем касаться именно работы с текстом, а не со звуком, потому что существуют уже неплохие готовые решения по переводу речи в текст (например, именно так изначально было записано то, что вы сейчас читаете :))

Что же касается бизнеса и NLP, тут дела обстоят очень даже хорошо. Считается, что цивилизация появляется, когда появляется письменность, и поэтому в цивилизованных компаниях уже давно принято вести документооборот на естественных языках. Также очень много текста логируется или записывается куда-то (например, тексты общения с оператором из клиентской поддержки), и оттуда можно достать много полезных данных.

Вообще практика вести документооборот очень старая, и поэтому у компании всегда есть какие-то накопленные документы и тексты, которые часто требуется обработать. Так что работа для NLP-инженера найдётся всегда.

Итак, давайте более конкретно поговорим о задачах, которые решает NLP-инженер.

### КЛАССИФИКАЦИЯ

Одна из самых популярных задач — это задача классификации текста. У нас есть набор документов, и каждый документ должен быть отнесён к одному из N заранее определённых классов. Хорошо знакомый вам пример — это спам-фильтр, который должен определить, является сообщение спамом или нет. Другой классический пример — это сентимент-анализ, или анализ тональности текста.  Например, у нас есть рецензии от КиноПоиска, и мы хотим понять по тексту, является эта рецензия позитивной, негативной или нейтральной.

### КЛАССИФИКАЦИЯ ТОКЕНОВ

Но кроме задачи классификации всего текста, иногда требуется выполнить задачу классификации токена.

Токен — это то, из чего состоят предложения (в русском и английском языке токены — это просто слова).

Например, задача «Определение частей речи в предложении» является задачей классификации токенов.

### NAMED-ENTITY RECOGNITION

Другая интересная задача — это Named-entity Recognition, или NER. В этой задаче мы пытаемся понять, является слово какой-то сущностью или нет.

Под сущностью может подразумеваться много чего. Чаще всего это персоны, географические объекты, названия каких-то учреждений или, например, даты, времена и валюты.

### ПОИСК

Другой крупный класс задач — это задачи, связанные с поиском и ранжированием.

Вероятно, все из нас на постоянной основе используют такие поисковики, как Яндекс или Google. В данном случае задача состоит в том, чтобы по запросу найти наиболее релевантные документы или, в случае Яндекса или Google, сайты, которые соответствуют этому запросу.



### СРАВНЕНИЕ ТЕКСТОВ

Также есть задача сравнения двух текстов. Например, для решения одной бизнес-задачи мне нужно было сопоставить названия товаров из двух разных баз данных двух различных магазинов. К примеру, в одном магазине товар назывался iPhone 10 XS Gold, а в другом магазине он назывался iPhone 10 золотой.

### ПОИСК ДУБЛИКАТОВ

Также существует задача поиска неполных дубликатов. Мне приходилось решать задачу, когда очень много сотрудников скидывали документы в файловые хранилища, и очень часто один и тот же документ лежал в разных местах. Проблема была в том, что появлялись разные версии документов, и наступала неразбериха. Задача была в том, чтобы во время того как закидывается документ, найти его копию или уже изменённую версию и предупредить об этом сотрудника.

### ЧАТ-БОТЫ И ДИАЛОГОВЫЕ СИСТЕМЫ

Ну и, конечно же, не стоит забывать об очень бурно развивающейся области чат-ботов и диалоговых систем, а также персональных помощников. Сейчас почти любая компания среднего размера пытается сделать собственного чат-бота для работы с клиентом или каких-то своих внутренних нужд: например, чтобы уменьшить нагрузку на операторов или автоматизировать процессы.

При этом сами чат-боты внутри могут быть совершенно разными. Они могут как просто реагировать на какие-то ключевые слова, могут работать на основании классификации намерений и специальных правил, которые говорят, что именно нужно ответить на конкретное сообщение, или же они могут самостоятельно генерировать ответ на совершенно любой вопрос.

### 3. Интересные задачи NLP
→ Но, конечно, только на классификации и поиске задачи в NLP не заканчиваются.

### KEYWORD EXTRACTION

Например, есть довольно интересная задача, которая называется Keyword Extraction.

У нас есть какой-то текст, и нам нужно выделить одно или несколько слов, которые максимально точно характеризуют этот текст.

Например, мы работаем с набором каких-то новостей. Используя Keyword Extraction, мы можем извлечь из них ключевые слова и сгруппировать их, исходя из наличия в них этих самых ключевых слов.

Многие из вас наверняка видели облако тегов: его как раз можно сгенерировать автоматически методами NLP.



### ВЕРИФИКАЦИЯ

Мы также можем использовать NLP для верификации пользователей. То есть по стилю сообщения и истории сообщений мы можем определять мошенников и понимать, пишет нам тот верифицированный пользователь, или же за него пишет кто-то другой.

### ASPECT-BASED SENTIMENT ANALYSIS

Другая интересная задача — это Aspect-based Sentiment Analysis. Она чем-то похожа на обычный анализ тональности текста, но с небольшим дополнением.

Представьте, что у нас есть отзыв на ресторан:

Еда была очень хорошая, но официанты были ужасно медленные и грубые. В целом интерьер в ресторане был нормальный.

Сложно понять, какая тональность у данного отзыва, потому что здесь есть три так называемых аспекта: «еда», «персонал» и «интерьер». И с разными аспектами у нас соотносятся разные настроения (сентименты): «еда» — позитивный, «персонал» — негативный, а «обстановка» — нейтральна. Этим и занимается Aspect-based Sentiment Analysis — определяет тональность разных аспектов в тексте. При этом в задаче могут быть заданы аспекты как заранее, так и может быть необходимо находить их автоматически.

### СУММАРИЗАЦИЯ ТЕКСТА

Представим, что вам прислали большое сообщение в мессенджере, и ваша задача — вывести это сообщение на экран умных часов. Но так как экран очень сильно ограничен по размеру, мы можем попробовать суммаризировать текст и вывести на экран только самое важное.

### 4. Инструменты NLP до нейронных сетей

→ Мы разобрали задачи, с которыми может столкнуться NLP-специалист. Теперь давайте посмотрим, как менялись инструменты, используемые в области NLP, в процессе её развития. Благодаря этим знаниям вы сможете понять, с какой стороны лучше всего подходить к решению той или иной задачи.

NLP тесно связано такой областью искусственного интеллекта, как **компьютерная лингвистика**. Можно даже сказать, что оно вышло из компьютерной лингвистики, или что NLP — это прикладная часть компьютерной лингвистики.

Что же такое компьютерная лингвистика? Если говорить простыми словами, она занимается математическим моделированием языка, а NLP занимается уже более прикладными вещами и гораздо чаще использует инструменты машинного обучения. Но, опять же, границы очень размыты.

Однако NLP не всегда основывалось на машинном обучении. В начале оно чаще использовало классические эвристические алгоритмы, когда находились какие-то закономерности в тексте и использовались для поставленных целей. Это могли быть анализ окончаний и приставок слов, взаимное расположение слов, наличие или отсутствие каких-то слов или подсчёт тех или иных метрик.

ПРИМЕР

Допустим, у нас есть задача поиска двух одинаковых слов. У нас есть какое-то эталонное слово, и есть набор других слов, с которыми мы будем его сравнивать. Но этот список слов может содержать слова, написанные с ошибками. То есть сравнение один-в-один тут не подойдёт, и для решения этой задачи нам достаточно применить так называемое **расстояние Левенштейна.**

**Расстояние Левенштейна** определяется как минимальное количество односимвольных операций (а именно вставки, удаления, замены), необходимых для превращения одной последовательности символов в другую.

После этого достаточно просто сказать, что если расстояние меньше, чем какое-то значение, то эти слова у нас одинаковые.



_Источник (https://habr.com/ru/post/114997/)_

Ну что же касается такой распространённый задачи, как классификация, то раньше, без машинного обучения, её решали довольно простыми методами. Например, для определения тональности текста мы можем просто говорить, что текст позитивный, если в нём содержатся слова «хорошо», «нравится», «отлично» в любых формах. И тональность текста негативная, если текст содержит слова «плохо», «не нравится», «ужас» и так далее.

Для поиска слов часто используются регулярные выражения.

В дальнейшем использовалась простая идея. У нас уже есть большое количество различных методов машинного обучения, которые работают на векторах. Логистическая регрессия, kNN, SVM. И если мы каким-то образом переведём текст в вектор, то сможем применять все уже известные нам методы.

Процесс перевода слова в вектор, а также сам результат этого перевода называется **эмбеддингом.**

### BAG-OF-WORDS

Самый простой способ перевода текста в вектор — **bag-of-words**, или **мешок слов.**

Назовём все слова, которые мы можем обработать, словарём, и представим, что у нас словаре всего 10 слов. Тогда для представления текста мы можем использовать вектор, где элемент вектора соотносится с каким-то словом, а значение элемента равно количеству вхождений этого слова в текст.

Допустим, у нас есть четыре  текста:

1. 'This is the first document.',
2. 'This document is the second document.'
3. 'And this is the third one.'
4. 'Is this the first document?'

Всего в них встречается 9 уникальных слов, если игнорировать большие буквы и пунктуацию. Наш словарь выглядит следующим образом:

['and', 'document', 'first', 'is', 'one', 'second', 'the', 'third', 'this']
И тогда для того чтобы перевести текст в вектор, мы используем вектор из 9 элементов: в нулевом элементе у нас будет значение, сколько раз в тексте встречается слово 'and', второй элемент — это количество слов 'document', и так далее.

a. 'This is the first document.'          [0 1 1 1 0 0 1 0 1]
b. 'This document is the second document.'	[0 2 0 1 0 1 1 0 1]
c. 'And this is the third one.'           [1 0 0 1 1 0 1 1 1]
d. 'Is this the first document?'          [0 1 1 1 0 0 1 0 1]
Например, посмотрим на векторизацию второго текста:

'This document is the second document.'	[0 2 0 1 0 1 1 0 1]
В тексте нет 'and', поэтому первый элемент равен 0, в тексте два раза встречается 'document', поэтому второй элемент равен 2 и т. д.

Ниже представлен код, с помощью которого вы можете получить bag-of-words.



```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
corpus = ['This is the first document.',
          'This is the second second document.',
          'And the third one.',
          'Is this the first document?',
         ]
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names() == (['and', 'document', 'first', 'is', 'one','second', 'the', 'third', 'this']))
```

```python
print(X.toarray())
```

```python
vectorizer.vocabulary_.get('first')
```

Объект, который создаёт мешок слов, называется **CountVectorizer**. Это трансформатор sklearn, на вход которому нужно передать корпус или список строк.

C помощью функции _get_feature_names()_  можно получить словарь.

Так как обычно после трансформации матрица содержит много нулей, vectorizer в результате возвращает не numpy array, a объект разрежённой матрицы. Но его можно трансформировать в array методом toarray.

А строчкой _vectorizer.vocabulary_.get('first') мы получаем индекс слова в словаре.

### TF-IDF

И, хотя мешок слов до сих пор применяется, есть гораздо более интересные методы. Например, метод **tf-idf** учитывает важность слов.

Сам метод довольно прост: для каждого слова в документах мы считаем его **tf — text frequency**. Это число показывает, сколько раз слово встретилось в документе. Конечно, это то же самое число, которое мы получали в мешке слов. Но дальше мы умножаем его на логарифм от **idf — inverse document frequency**, где idf — это количество всех документов, поделённое на количество документов, в которых встречается данное слово.

Для чего всё это? Например, если слово встречается довольно часто во всех документах, то есть, скорее всего, является каким-то очень часто применяемым в языке словом вроде артикля или предлога, то idf такого слова будет близок к единице — значит, это слово явно не очень хорошо характеризует документ.  Но если слово встречается в малом количестве документов, то idf будет большим числом, и мы таким образом как бы усиливаем вес слова.

Давайте теперь посмотрим, как можно программно получить tf-idf-матрицу. Для этого сначала нужно посчитать tf с помощью CountVectorizer.

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
corpus = [
   'This is the first document.',
   'This is the second second document.',
   'And the third one.',
   'Is this the first document?',
]
X = vectorizer.fit_transform(corpus)
X = X.toarray()
vectorizer.get_feature_names()
```

И теперь можно посчитать tfidf:



```python
from sklearn.feature_extraction.text import TfidfTransformer
import numpy as np
transformer = TfidfTransformer(smooth_idf=False)
tfidf_matrix = transformer.fit_transform(X)
transformer.idf_
```

```python
tfidf_matrix
```

На пятой строчке мы вывели idf. Значения в этом векторе соотносятся со словами, полученными vectorizer.get_feature_names().

Самые большие значения idf у слов ['and', 'one', 'second', 'third'], так как они встречаются только в одном документе, а самое маленькое значение — у слова  'the', так как оно встречается во всех документах.

Значения в tf-idf-матрице нормализованы, и поэтому сложно сказать, правильно мы посчитали значение выше. Но можно заметить, что второй элемент в transformer.idf-векторе отвечает за слово 'document'. И так как tf у этого слова равен 1, то значение idf слова 'document' также равно и tf-idf. Следовательно, значение idf равно тому значению, что мы посчитали.

У методов bag-of-words и tf-idf есть несколько минусов:

- В реальной задаче у нас будет не десяток слов в словаре, а тысячи.
- Документы не всегда могут быть большими, а для маленьких предложений векторы в основном будут содержать нули.
- Мы не учитываем взаимное расположение слов: если нам дать предложение ‘snakes eat fish’, то, переведя его в вектор, мы никогда не узнаем, кто кого ест.

Дальше в NLP решили, что, раз уж у нас есть методы машинного обучения, так давайте их и применим для того, чтобы получить из слов векторы. То есть мы из слова пытаемся получить вектор в каком-то смысловом пространстве с помощью методов ML. Векторы слов, которые схожи по смыслу, должны быть рядом в этом пространстве.

### WORD2VEC

Один из первых популярных методов назывался **GloVe**. Он использовал статистические методы для получения векторов из слов. А немного позднее появился другой популярный метод, основанный на плоской нейронной сети, — Word2Vec.

Работает Word2Vec довольно просто:

- Берём корпус текста: это может быть и текст Википедии, и какой-то художественный рассказ.
- Создаём огромную матрицу размером количество слов x количество пространства смыслового вектора, и заполняем её случайными значениями. Эта матрица содержит строки с векторным представлением слова.
- Обучаем нейронную сеть: берём слово из предложения, подаём его на вход нейронной сети и просим её предсказать, какие другие слова были рядом с этим словом, то есть просим её предсказать контекст слова. Или же мы можем сделать наоборот: берём контекст слова и просим нейронную сеть предсказать, для какого слова мы взяли контекст.

С помощью Word2Vec авторы метода добились невероятного результата.

Например, если мы возьмём вектор слова King, вычтем из него вектор слова man и прибавим вектор слова woman, в результате мы получим вектор, который очень близок к вектору слова queen.

### FASTTEXT

И хотя Word2Vec до сих пор применяется, у него есть один очень большой минус: _Word2Vec не способен сгенерировать вектор для слова, которое отсутствовало в обучающем датасете._

Именно поэтому авторы Word2Vec предложили новый метод, который называется **FastText**. В отличие от Word2Vec, FastText учится создавать векторы для частей слова (charagram).

Допустим, у нас есть слово «енот». При обучении FastText разобьёт его на части: енот - (ен, но, от, ено, нот). И векторы внутри FastText при обучении будут создаваться для каждой части слова, а не для целого слова. И чтобы получить конечный вектор слова «енот», FastText просто сложит векторы его частей.

И получается, что, когда на вход передаётся слово, которого не было в обучающей выборке, FastText делит его на части и уже на их основании получает вектор слова. Данный алгоритм отлично работает для русского и английского языков.
