### ЧТО ВХОДИТ В АРСЕНАЛ NLP-ИНЖЕНЕРА?

В качестве одного из основных инструментов естественно используется какой-то язык программирования, чаще всего это Python или R. В данном курсе мы будем сосредоточены только на Python, так как R больше используется в академических кругах.

В качестве основной библиотеки Python для NLP чаще всего используется **NLTK (Natural Language Toolkit)**. Это, наверное, один из самых крупных модулей по работе с текстом, особенно с английским языком. Также есть **SpaCy**, который также имеет схожий арсенал по обработке текста. Эти библиотеки содержат в себе огромные возможности по обработке, трансформации и анализу текста. Другой популярный инструмент — это **Gensim**, он уже в себе содержит некоторые модели для перевода текста в векторное представление, о котором мы поговорим далее.

### 2. Классические задачи NLP

→ Для начала разберёмся с тем, какие задачи решает NLP-инженер.

Вообще задачи в NLP можно разделить на две большие группы:

- Работа  со звуком (когда мы работаем с записанной речью).
- Работа с текстом.

Но для работы со звуком мы просто можем перевести речь в текст методами Speech2Text и затем применить методы обработки текстов. И в курсе в основном мы будем касаться именно работы с текстом, а не со звуком, потому что существуют уже неплохие готовые решения по переводу речи в текст (например, именно так изначально было записано то, что вы сейчас читаете :))

Что же касается бизнеса и NLP, тут дела обстоят очень даже хорошо. Считается, что цивилизация появляется, когда появляется письменность, и поэтому в цивилизованных компаниях уже давно принято вести документооборот на естественных языках. Также очень много текста логируется или записывается куда-то (например, тексты общения с оператором из клиентской поддержки), и оттуда можно достать много полезных данных.

Вообще практика вести документооборот очень старая, и поэтому у компании всегда есть какие-то накопленные документы и тексты, которые часто требуется обработать. Так что работа для NLP-инженера найдётся всегда.

Итак, давайте более конкретно поговорим о задачах, которые решает NLP-инженер.

### КЛАССИФИКАЦИЯ

Одна из самых популярных задач — это задача классификации текста. У нас есть набор документов, и каждый документ должен быть отнесён к одному из N заранее определённых классов. Хорошо знакомый вам пример — это спам-фильтр, который должен определить, является сообщение спамом или нет. Другой классический пример — это сентимент-анализ, или анализ тональности текста.  Например, у нас есть рецензии от КиноПоиска, и мы хотим понять по тексту, является эта рецензия позитивной, негативной или нейтральной.

### КЛАССИФИКАЦИЯ ТОКЕНОВ

Но кроме задачи классификации всего текста, иногда требуется выполнить задачу классификации токена.

Токен — это то, из чего состоят предложения (в русском и английском языке токены — это просто слова).

Например, задача «Определение частей речи в предложении» является задачей классификации токенов.
