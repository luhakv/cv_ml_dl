Что вообще означает NLP? Не стоит путать NLP с другой, многим знакомой аббревиатурой НЛП (нейролингвистическое программирование).

NLP расшифровывается как **Natural Language Processing (обработка естественного языка)**. Как понятно из названия, основная область работы NLP — это естественный человеческий язык. «Естественный» означает, что именно на этом языке мы с вами говорим и пишем без каких-либо ограничений.

Комментарий от автора

У NLP есть интересная особенность. В компьютерном зрении неважно, где сфотографировали котика, в Африке или в России — он всегда остаётся котиком. Для NLP же часто приходится создавать решение под каждый язык, так как грамматика и слова на разных языках очень сильно отличаются, и часто решения, которые хорошо работают для одного языка, могут не заработать для другого, особенно если он из другой языковой семьи.

### ЧТО ВХОДИТ В АРСЕНАЛ NLP-ИНЖЕНЕРА?

В качестве одного из основных инструментов естественно используется какой-то язык программирования, чаще всего это Python или R. В данном курсе мы будем сосредоточены только на Python, так как R больше используется в академических кругах.

В качестве основной библиотеки Python для NLP чаще всего используется **NLTK (Natural Language Toolkit)**. Это, наверное, один из самых крупных модулей по работе с текстом, особенно с английским языком. Также есть **SpaCy**, который также имеет схожий арсенал по обработке текста. Эти библиотеки содержат в себе огромные возможности по обработке, трансформации и анализу текста. Другой популярный инструмент — это **Gensim**, он уже в себе содержит некоторые модели для перевода текста в векторное представление, о котором мы поговорим далее.

_Источник (https://www.machinelearningmastery.ru/named-entity-recognition-3fad3f53c91e/)_

И, конечно, не стоит забывать о библиотеках, которые создавались для русского языка. Это такие проекты, как **Natasha** и **DeepPavlov** из МФТИ.

_Источник (https://deeppavlov.ai/)_

И если раньше чаще применяли методы, основанные на каких-то эвристиках, то в последние 10-20 лет машинное обучение приобрело невероятную популярность в NLP. И поэтому в список инструментов NLP инженера входят также стандартные библиотеки и фреймворки машинного и глубокого обучения, например, scikit-learn, PyTorch, Tensorflow2/Keras и так далее.

В этом модуле мы:

- рассмотрим историю развития NLP, а также посмотрим на библиотеки, которые применяются для различных задач;
- разберём, как можно решить задачи без машинного обучения стандартными средствами и эвристическими алгоритмами (что до сих пор довольно часто применяется) ;
- посмотрим несколько способов получения векторов из естественного текста и слов, а также обсудим особый вид нейронных сетей под названием рекуррентные нейронные сети, который очень часто применяется в NLP и в работе с последовательностями;
- конечно же, попробуем решить некоторые задачи этими методами, чтобы лучше понять, чем занимается NLP-инженер;
- а в самом конце мы обсудим трансформеры. И нет, это не роботы, которые могут превращаться в машины. Это особая нейронная сеть, которая появилась сравнительно недавно и которая сейчас является, наверное, самым мощным инструментом для обработки естественного языка.
