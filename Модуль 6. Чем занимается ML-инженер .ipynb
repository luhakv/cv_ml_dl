### 2. Кто такой ML-инженер, откуда и куда он идёт?
ML Engineer — это специалист по полному циклу разработки моделей машинного обучения, начиная со сбора данных и построения моделей и заканчивая написанием бэкенд-сервисов и анализом экспериментов. В этой специальности важен хороший уровень в Computer Science и знание алгоритмов машинного обучения.

В разных компаниях ML-инженеры ведут себя по-разному:

**В маленьких компаниях** это может быть Full-Stack ML Engineer, который может и найти данные, и проанализировать, и сделать модель, и выложить её на сервер. Иными словами, он может сделать вообще всё, при этом не обладая специализацией в какой-то конкретной области.	

**В крупных компаниях** это может быть более узкий специалист, который занимается каким-то конкретным набором задач, на которых он специализируется. Например, построением процесса обработки данных на кластере.
Выполняемые рабочие задачи зависят не только от того, в какой компании вы работаете, но и от того, из какой области вы пришли в ML-инженерию. 

В Data Science (в том числе в ML Engineering) можно прийти из любой области и найти способы применения своего предыдущего опыта к новым задачам.

### 3. Бизнес-задачи ML-инженера

→ Список задач у ML-инженера может быть бесконечным — Data Science, похоже, можно применить где угодно, и эта мысль только подтверждается опытом.

Несмотря на это, есть набор типичных задач.

### ОТТОК КЛИЕНТОВ

Считается, что удержать клиента проще, чем привлечь его. Знание о том, что клиент собирается от нас уйти, — это очень полезная информация в мире высокой конкуренции и ограниченной аудитории потенциальных клиентов. Поэтому мы предпримем какие-то действия, чтобы клиент остался с нами. Например, дадим ему скидку на наши продукты.

→ Как получить эту информацию? Есть модели, которые могут это предсказать. Основные задачи ML-инженера во всех случаях одни и те же: построить модель, сделать так, чтобы ей можно было пользоваться. 

### СКОРИНГ КЛИЕНТОВ

Ещё одной важной задачей является скоринг, например кредитный скоринг — это когда на основе вашей кредитной истории формируется рейтинг, а уже на его основании принимается решение, под какими условиями вам дадут кредит и дадут ли вообще.

### РЕКОМЕНДАТЕЛЬНЫЕ СИСТЕМЫ

Рекомендательные системы повсеместно используются в интернет-магазинах. Вы заходите на сайт, выбираете товар, а вам предлагают похожие товары или такие товары, которые заказывали похожие на вас люди. 

→ Как это работает? Модель определяет ваш демографический портрет и смотрит на историю ваших запросов. Затем она находит похожих на вас людей и смотрит, что они искали в магазине. Эти же товары покажут и вам.

→ Кто создаёт эти модели? ML-инженер!

### ЦЕНООБРАЗОВАНИЕ

В крупных торговых сетях очень много продуктов. Цена на эти продукты может зависеть от многих факторов: географического положения магазина, демографического портрета типичного покупателя, конкуренции в регионе, скорости поставок и многого другого.

### 4. Технические задачи ML-инженера


Как мы уже обсуждали, у ML-инженера в ходе разработки появляется много разных задач. В разных командах и у разных специалистов список задач может отличаться. 

Кроме того, отличаются задачи на разных этапах разработки:

**До работы с моделью**
- Сбор данных.
- Анализ, очистка данных и создание датасетов для моделей. Этот процесс известен как EDA (Exploratory Data Analysis, разведывательный анализ данных).

**Во время работы с моделью**
- Поиск подходящих моделей для решения задачи.
- Обучение моделей.

**После работы с моделью**
- «Выкатка» моделей в продакшен в бэкенд-сервисах и мобильных клиентах.
- Создание и сбор метрик, проведение экспериментов.

**TL (Extract, Transform, Load)**
- Задачи, которые проходят сквозь весь процесс разработки. Этот набор связан со сбором, хранением и обработкой данных.



Говоря простым языком, ML-инженер должен организовать систему, с помощью которой можно будет:

- Удобно выгружать откуда-то данные (Extract). 
- Иметь возможность применить разные трансформации к данным (Transform).
- Загружать изменённые данные в базы данных, чтобы затем обучать на них модели (Load).  

Эта система значительно облегчает моделирование, так как она позволяет иметь легкий доступ к нашим данным и способам работы с ними.

### 5. Модели и бизнес


→ Итак, вам выдали модель машинного обучения. Она выдаёт на выходе предсказания стоимости домов, вы можете оценивать их с помощью метрик качества. 

Что ж, вы можете нарисовать графики изменения стоимости в зависимости от количества комнат в доме и сделать выводы на основе данных. Но что дальше?

**Важно задаться следующими вопросами:**

- Кто будет основными потребителями результатов: другие сервисы? Или модель будет встроена в мобильное приложение?
- Как и в каком виде потребители ожидают получать результаты? По времени или по запросу? И если по запросу, то по каким правилам они будут осуществлять запрос?
- Какая существует или планируется инфраструктура для этого?


### ДЕПЛОЙ

Задача развёртывания приложения на сервере называется деплой (от англ. deploy, deployment — внедрение, развёртывание). По сути это процесс трансформации исходного кода вашего приложения в рабочее состояние на конкретном сервере. 

Проще говоря, деплой — это когда вы берёте модель и помещаете её на сервер, чтобы ей могли пользоваться другие.

Для того, чтобы внедрить модель в продакшен, необходимо сохранить обученную модель в файл, поднять сервер, а затем доставить и запустить на нём свою модель. Звучит довольно просто, не так ли?

### ВСЁ НЕ ТАК УЖ ПРОСТО!

К сожалению, при моделировании зачастую не учитываются вопросы, связанные с деплоем модели и её внедрением в уже существующие системы. 

→ Для чего мы обучили модель? Может быть, мы будем использовать её для чат-бота? Может быть, она должна быть встроена в приложение для iPhone? Как часто потребуется переобучать модель?

Опыт подсказывает, что если модель создаётся «в вакууме», то на выходе она просто не сможет быть доставлена в продакшен. 

Дополнительной проблемой является и тот факт, что большинство библиотек для машинного обучения сфокусированы именно на обучении, а не на предсказании. 

Ещё одна проблема: в крупных компаниях дата-сайентисты, дата-инженеры, ML-инженеры, занимающиеся внедрением моделей в продакшен, могут быть не просто разными людьми, но и разными командами. На практике может оказаться так, что готовая обученная модель будет внедрена другой командой на другом языке программирования. При этом обычным разработчикам модели машинного обучения могут представляться непрозрачными тёмными ящиками, которые работают, пожалуй, только с помощью чёрной магии. В общем-то, так и есть.  

Наконец, если вы сделаете модель и она будет работать, допустим, с точностью 95 % — это хорошо или плохо? Может показаться, что всегда можно лучше. 

Давайте потратим ещё месяц и добъёмся точности 96%. Потратим ещё три — и точность будет 97 %. Но действительно ли нам нужно тратить дополнительное время? Обычно требования диктует бизнес, и бизнесу важна скорость. А мы далеко не всегда видим значительное увеличение прибыли при небольшом изменении точности.

Поэтому важно запомнить очень важную мысль: 

### МОДЕЛИ В ВАКУУМЕ НИКОМУ НЕ НУЖНЫ 

Модель приносит пользу только тогда, когда ей пользуются и когда это соответствует требованиям бизнеса.

### МОДЕЛЬ — ЭТО СЕРВИС

Интересное следствие из предыдущей мысли: модель — это сервис.

Почему так?

- Модели в вакууме никому не нужны.
- Модели нужны нам для решения конкретных задач.
- Модели часто являются всего лишь кусочком единой системы.
- Эта система обращается к модели и ожидает от неё конкретный результат.  

Это означает, что мы должны не просто сделать модель, но ещё и выложить её куда-то и предоставить интерфейс взаимодействия с ней. Об этом мы поговорим совсем скоро.

### 6. Жизненный цикл ML-проекта

Мы поговорили о задачах, которые решает ML-инженер. А как он это делает в реальных компаниях?

### 7. Что такое воспроизводимость?


Если сравнивать рабочий процесс обычного разработчика ПО и специалиста по анализу данных, можно найти ряд принципиальных различий. Код, разрабатываемый дата-сайентистом, должен быть воспроизводим в постоянно меняющихся в условиях:

- Меняются данные, поступающие на вход.
- Трансформируются пайплайны препроцессинга данных.
- Постоянно изменяются гиперпараметры алгоритмов.  

Ожидается, что выполнение кода должно приводить к одинаковым результатам в различных условиях. Воспроизводимость результатов является одним из главных показателей качества вашей модели. Поэтому обеспечение воспроизводимости является одной из важнейших проблем в современном ML-инжиниринге.

Мы можем доверять результату, полученному с помощью модели, только в том случае, если точно известно, на какой выборке и с каким пайплайном предобработки она была построена. Кроме того, ожидается, что результаты будут консистентными, так как нужно понимать, какое изменение привело к улучшению или ухудшению результатов.

Когда мы работаем над моделью и пишем код, мы проходим через все этапы на изображении выше. Чтобы не запутаться в том, что мы пишем, и упростить разработку в команде, мы используем системы **контроля версий**, например git, с которым вы уже знакомы.

В процессе работы над проектом порождаются различные артефакты: датасеты, модели, файлы конфигураций и прочее. Для их **версионирования** обычно используются такие инструменты, как DVC (https://dvc.org/) и Nexus Sonatype (https://www.sonatype.com/products/repository-pro).

При этом представление модели и данных внутри программы и на диске компьютера могут различаться. Для того, чтобы мы могли не только обучить модель, но и сохранить её, а потом загрузить, используется **сериализация**. Про неё мы поговорим чуть позже.

### НАСТРОЙКА ОКРУЖЕНИЯ И СРЕДЫ ВЫПОЛНЕНИЯ

Во время работы и исполнения кода нужно обеспечить одинаковые ОС, установить все зависимости и одинаковые версии библиотек. Чтобы среда выполнения была одинаковой, мы используем инструменты изоляции и контейнеризации, например **Docker** (https://www.docker.com/). 

Docker позволяет «заворачивать» модель в контейнер со всеми зависимостями, которые ей нужны. После этого контейнер можно запустить где угодно и быть уверенным в том, что он сработает как надо.

Docker — одно из средств **виртуализации**. Говоря простым языком, виртуализация — это запуск одной или нескольких систем в рамках другой операционной системы на одной вычислительной машине.

Если есть несколько моделей от разных команд разработки, и эти модели использовали разные версии библиотек или даже разные языки программирования, мы можем запустить их в разных контейнерах на одном компьютере и обращаться к ним из какой-то другой программы. 

Представим, что мы — Amazon, и у нас есть несколько моделей машинного обучения: рекомендательная система, скоринг клиентов, оптимизация нагрузки на базу данных. Мы можем запустить эти модели в разных контейнерах и обращаться к ним с единого сервера, на котором поднят наш сайт.

Используя контейнеризацию таким образом, мы поддерживаем в нашем проекте **изолированность**: приложение будет запускаться в своей отдельной среде и таким образом не зависеть от ОС и настроек системы, на которой мы разворачиваем приложение. Это является одним из этапов настройки виртуального окружения.

### 8. Работа с IDE и контролем версий


→ По большей части все задачи ML-инженера связаны с написанием кода. Где же писать код? 

На самом деле можно это делать где угодно, хоть в блокноте.

Но программисты — ленивые люди, которые стремятся достичь результата за меньшее количество действий. Поэтому используются **IDE** — интегрированные среды разработки (от англ. _Integrated Development Environment_).

**IDE** — это среда разработки для программистов, которая помогает автоматизировать рутинную деятельность и вообще делает программирование намного удобнее и комфортнее для самого программиста. В ней возможно применить автозаполнение, проверки, отладку кода и даже более сложные вещи — написание целого блока кода нажатием двух клавиш.

В Data Science в основном используется два типа IDE: классические и основанные на ноутбуках. Давайте разберём плюсы и минусы каждой из них.

### Классические IDE



Примерами классических IDE являются **PyCharm** (есть платная и бесплатная версии) и **Visual Studio Code**, или **VSCode** (бесплатная IDE, https://code.visualstudio.com/).

⭐ Совет от автора
Рекомендую использовать VSCode — этот инструмент бесплатный, удобный и очень гибко настраивается плагинами.

VSCode отлично подходит для написания проектов, состоящих из множества файлов. В нём удобная навигация, много возможностей по настройке с помощью плагинов.

Если говорить совсем простым языком, то VSCode (и любая другая классическая IDE) — это просто текстовый редактор с кучей «наворотов» специально для написания кода.

Например, вы можете написать код сервера, сохранить его и запустить в консоли. Как раз для этого мы и будем использовать классические IDE дальше.

### IDE, основанные на ноутбуках



IDE, основанные на ноутбуках, предлагают другую парадигму написания кода. В вашем распоряжении — не единое окно с текстовым редактором, а набор ячеек. 

Вы можете написать код в ячейке, выполнить его, перейти к другой ячейке и работать дальше в ней. При этом, результаты из первой ячейки будут сохранены в памяти.

Почему это удобно? Такой способ работы сильно облегчает анализ данных, потому что есть возможность быстро проверить свою идею с данными в ячейке, поменять какие-то параметры и снова проверить. А в другой ячейке вас уже будет ждать код, который использует эти данные дальше.

Если у вас большой проект, то вам не придётся ждать, пока весь код выполнится, чтобы получить результат — достаточно выполнить код всего в одной ячейке. 

В то же время запустить сервер в ноутбуке будет сложнее: как только вы выполните ячейку с запуском сервера, весь ноутбук как бы «зависнет», ожидая завершения работы этой ячейки. Но работа завершаться не будет, ведь сервер будет продолжать работать.

Самой популярной средой для работы с ноутбуками является **Jupyter Notebook**, с которым вы уже работали.

### КАКУЮ IDE ВЫБРАТЬ?

Какую IDE выбрать? Это зависит от ситуации. Если мы занимаемся анализом данных (например, загрузкой данных из файла и дальнейшей обработкой этих данных с целью найти интересные зависимости), то лучше использовать ноутбуки.

Если мы уже обучили модель и готовимся к деплою, лучше использовать классические IDE.

### НЕМНОГО О КОНТРОЛЕ ВЕРСИЙ

Независимо от типа IDE, мы всё равно будем пользоваться контролем версий.

Контроль версий нужен, чтобы работа над кодом была удобнее как в команде, так и при работе над проектом в одиночку.

Вы когда-нибудь писали код? Или может диплом? 

Возможно, вы встречались с проблемой, когда в папке лежат несколько файлов с названиями вроде Диплом 2020 final финальная v1.5.



Как сохранить стабильную версию файла? Что если вы сохраните изменения, но вам потребуется вернуться к предыдущей версии? Как разобраться в этих версиях, если их очень много?

Даже если вы работаете в одиночку, то можете столкнуться с этими проблемами. Если же вы — часть команды, то эти вопросы просто необходимо как-то решать.

Поэтому придумали такую вещь, как системы контроля версий. Это программы разной степени сложности. Есть такие, которые просто хранят разные версии файлов у вас на компьютере. Но таких приложений недостаточно для совместной работы над множеством файлов с необходимостью проверки и исправления кода в огромном количестве.

**Системы контроля версий** — это более сложные программные решения, которые позволяют одновременно работать над файлами множеству людей, сохранять версии, откатывать изменения, делать код-ревью (проверку вашего кода другими разработчиками) и многое другое. 

Существуют различные системы для контроля версий: _SVN, Mercurial, Perforce, CVS, Bitkeeper_ и другие. Но самой популярной является **Git**.

Благодаря тому, что Git распределённый, можно вносить изменения в код без доступа к интернету, локально, сохраняя все данные в папках на жёстком диске, которые называются **репозиторием.** 

Чтобы иметь доступ к репозиторию с различных устройств, можно хранить копию репозитория онлайн, что сильно облегчает работу над одним проектом для нескольких людей. Для этой цели существуют специальные сайты, такие как _github, bitbucket или gitlab_.

### 9. Поговорим о тестировании


→ Чем сложнее система, тем больше шансов допустить ошибку. Поэтому перед выпуском программного продукта его нужно протестировать. 

Есть такой термин — **пирамида тестирования**. Это означает, что есть тесты, находящиеся на разных уровнях абстракции и, следовательно, проверяющие систему по-разному.

Схема ниже — интерактивная. Нажмите на значок «руки» в правом верхнем углу — он укажет вам все элементы, на которые можно кликнуть ↓

https://view.genial.ly/60e85ca0a697cb0d91305585/horizontal-infographic-diagrams-dsprprofml4

1) **Юнит-тест (Unit-тест)** — модульное тестирование, проверка отдельных модулей кода, набора модулей с данными, процедурами использования и обработки. Тестируют одну небольшую единицу функциональности. Тесты пишутся отдельно для каждой необычной функции или метода. Как можно протестировать работоспособность модели? Выбрать архитектуру и готовые хорошие данные, проверить, что выход модели соответствует ожиданиям.

2) **Integration test (Интеграционный тест)** — тестирование модулей после unit-теста, проверка взаимодействия между разными модулями или единицами функциональности. Например, проверка, что после загрузки данные корректно поступают на вход методам предобработки данных, затем корректно поступают на вход в модель. Это тестирование всего **Pipeline — процесса обработки данных и обучения модели.**

3) **End-to-end Функциональный тест** — тестирование системы на соответствие функциональным требованиям. На этом этапе мы пытаемся определить, действительно ли система делает то, что от неё хотят. В отличие от предыдущих тестов, которые пишут разработчики, функциональные тесты пишут тестировщики. Предполагается, что разработчики знают устройство системы, а тестировщики не знают, поэтому при написании могут полагаться только на требования, которые были поставлены к системе на этапе Business Understanding.

### 10. Сохранение и загрузка моделей

→ Что из себя представляет обученная модель?

По своей сути, как и всё (ну, почти всё) в языке программирования Python, она является объектом. Этот объект не является простым, ведь ваша модель содержит сложную иерархию классов; в каждом классе есть набор полей, ссылающихся на объекты других классов и так далее.


Чтобы гарантировать сохранение всей структуры данных и получить её при загрузке обратно, используется сериализация.

**Сериализация** — процесс трансформации любой структуры данных, поддерживаемой в языке, в последовательность битов. Обратной к операции сериализации является операция десериализации.

→ ПРИМЕР

Приведём простой пример с файлом формата CSV. Пусть у нас есть файл с одной строкой, где несколько слов записаны через запятую. На диске это будет записано просто как последовательность байтов.

Однако когда мы считываем её внутри программы, мы хотим работать с этой строкой как со списком строк, то есть нам нужно не только считать эти данные, но и применить к ним некоторое преобразование:

```python
line = 'word1, word2, word3'
line.split(",")
```

Можно заметить, что после этого преобразования у нас появилось много объектов: каждое слово списка и некоторая структура — то, как слова лежат в этом списке. Процесс, который мы только что выполнили, называется десериализацией.

Также возможен обратный процесс, когда мы хотим сохранить в виде последовательности байтов какие-то данные из памяти программы. Например, нам нужно снова получить начальную строку:

```python
",".join(['word1', 'word2', 'word3'])
```

Этот процесс называется **сериализацией.**

### PICKLE



Посмотрим, с помощью каких средств происходит сериализация объектов в Python.

В стандартную библиотеку Python входит модуль _Pickle_ (https://docs.python.org/3/library/pickle.html), который служит для сериализации почти всех объектов произвольного типа. 

Мы помним, что объекты находятся в оперативной памяти и направляются в байтовые потоки ввода-вывода. В байтовые потоки может быть направлен любой файлоподобный объект. 

В ходе десериализации исходный объект воссоздаётся в оперативной памяти с теми же самыми значениями, но с новой идентичностью — новым адресом в памяти.

⛔ **Внимание!**

Так как законсервирован может быть абсолютно любой объект, то в нём могут быть «спрятаны» различные вредоносные программы или данные. Поэтому будьте внимательны и не проводите десериализацию бинарных файлов, в происхождении которых вы не уверены.

Более того, обработку данных и моделирование в Python можно включать в единый пайплайн. Простыми словами, это список объектов, которые что-то делают с данными, например масштабирование признаков; выбор подходящих признаков; модель линейной регрессии. С пайплайном можно обращаться как с обычным объектом и его точно так же можно сериализовать, как и просто модель.

→ Мы потренируемся в сериализации и десериализации позже в этом модуле.

### ФОРМАТЫ PMML И ONNX-ML

Среда или требования к **инференсу** (той задачи, для которой необходима обученная модель) модели для вашего проекта могут быть устроены так, что потребуют реализации на другом языке программирования, отличном от Python. Например, если компания разрабатывает десктопное приложение, то для внедрения модели её потребуется «перевести» на Java или C++. Как это сделать?

В таких случаях используется генерация файла формата _PMML (Predictive Model Markup Language)_. 

**Определение: PMML

PMML** — это XML-диалект, который используется для описания статистических моделей и моделей data science. PMML-совместимые приложения позволяют легко обмениваться моделями данных между собой. Разработка и внедрение PMML осуществляется IT-консорциумом Data Mining Group (http://dmg.org/pmml/v4-4/GeneralStructure.html).

С помощью этого инструмента можно сделать так, чтобы в одном файле содержалась вся информация для того, чтобы модель или пайплайн могли быть использованы на любом языке программирования.

В разработке моделей на основе нейронных сетей сегодня наиболее распространён формат _ONNX (Open Neural Network Exchange)_.

**Определение: ONNX**

**ONNX (Open Neural Network Exchange, читается как «оникс» - https://onnx.ai/)** — это открытый стандарт для обеспечения совместимости моделей машинного обучения. Он позволяет разработчикам искусственного интеллекта использовать модели с различными инфраструктурами, инструментами, средами исполнения и компиляторами.

Стандарт поддерживается совместно компаниями Microsoft, Amazon, Facebook и другими партнёрами как проект с открытым исходным кодом.

→ Информация на будущее

Часто стандарт ONNX и его библиотеки используют для конвертации из одного фреймворка в другой (например, из Pytorch в Tensorflow для использования в продакшне). Для конвертации различных фреймворков (не только DL) в формат ONNX и обратно существует ряд библиотек (https://github.com/onnx): 

ONNX-Tensorflow (https://github.com/onnx/onnx-tensorflow)  
Tensorflow-ONNX (https://github.com/onnx/tensorflow-onnx)  
Keras-ONNX (https://github.com/onnx/keras-onnx)  
Sklearn-ONNX (https://github.com/onnx/sklearn-onnx)  
…и другие.    

Также в рамках стандарта ONNX есть инструмент ONNX-runtime (https://github.com/microsoft/onnxruntime). Он служит для ускорения инференса Python-моделей, а также инференса на других языках, например Java, C++.

### 11. Офлайн- и онлайн-модели

В самом начале мы говорили о том, что модель в вакууме никому не нужна — она приносит пользу только если встроена в какой-то бизнес-процесс. От того, как выглядит этот процесс, зависит и то, как нужно встраивать туда модель. Давайте разберём самые частые примеры. 

**Принципиально модели можно разделить на два типа: офлайн и онлайн.**

**Определения: онлайн- и офлайн-модель**

**Онлайн-модель**, или модель, доступная в реальном времени, подразумевает, что существует некий постоянный процесс с этой моделью. К этому процессу можно делать запросы, чтобы получить предсказание модели. Эти запросы могут делать люди, а могут делать и другие процессы. Важно, что в такой схеме предсказание можно получить сразу, как только инициатору запроса это требуется.  

**Офлайн-схема** работает иначе: модель не доступна для запросов, а наоборот, запускается один раз на всех доступных на текущий момент данных и может делать это с некоторой периодичностью. 

Давайте разберём несколько примеров в порядке увеличения автоматизации.

### СОРЕВНОВАНИЯ НА KAGGLE

Самый простой пример офлайн-модели, которая не требует никакой автоматизации — соревнования по анализу данных. Действительно, оставляя за скобками само построение модели, когда она уже есть у вас на руках, вы один раз применяете её ко всем тестовым данным и заливаете очередной сабмит с результатами. Тут нечего автоматизировать, потому что модель применяется только один раз.

### Кейс. Рекомендации для небольшого онлайн-магазина

Итак, нам необходимо запускать скрипт каждый день в определённое время. Пусть это будет 12 часов вечера, когда активности уже нет.

Но если это произошло, не всё потеряно: есть свои планировщики и в Windows (https://clck.ru/Ww4J5), либо можно отдельно установить cron, используя cygwin (https://habr.com/ru/post/149545/).

### 12. Кратко о сервисах

→ Итак, мы поговорили о типах моделей и затронули вопрос о том, что модель — это сервис. Давайте обсудим это подробнее.

Требования продакшн-среды несомненно влияют и на выбор сервисной архитектуры.

Существует два основных типа: **монолитная** и **микросервисная архитектуры.**


**Монолитная**  
Все контексты приложения запускаются на сервере с помощью внутрипроцессных взаимодействий. То есть приложение включает в себя все функции, и они взаимодействуют в одной запущенной программе.


**Микросервисная**  
Подход, позволяющий инкапсулировать определённые контексты приложения. Таким образом, каждая функция вынесена в отдельную небольшую программу или сервис.


### ПРЕИМУЩЕСТВА МИКРОСЕРВИСНОГО ПОДХОДА В РАЗРАБОТКЕ

✔ Мы повышаем автономность различных частей продукта. Мы можем разрабатывать их раздельно и разными командами (отслеживая при этом версионность!)

✔ Мы обеспечиваем гибкую масштабируемость. За счёт модульности приложения мы можем масштабировать только ту часть, которая требует дополнительных ресурсов.

✔ Мы повышаем стабильность. Если что-то упало, то будет недоступен только один функционал, а не всё приложение. 

✔ Наконец, микросервисы позволяют запускать приложение на разных серверах, что обеспечивает горизонтальную масштабируемость.

### МАСШТАБИРУЕМОСТЬ СИСТЕМЫ

- Что делать с тяжёлыми запросами к нашим данным, которые бегут часами? 
- И так далее и тому подобное…  

Обобщая эти и подобные вопросы, говорят о масштабируемости системы — способности системы выдерживать рост нагрузки по мере добавления ресурсов.

Мы поговорили о сервисах. Но давайте вспомним, как они общаются между собой.

Общение всегда происходит по какому-то **протоколу. Протокол** — это некоторый набор правил. В нашем случае это правила, по которым программа, получив по сети набор битов, понимает, как его прочитать и что он значит. 

### HTTP

**HTTP** — это самый широко используемый протокол. Все сайты, на которые вы заходите, работают по этому протоколу. Изначально он и был разработан именно для передачи содержимого HTML-страниц во всемирной паутине, но впоследствии стал использоваться и в других целях. Например, HTTP используется для налаживания взаимодействия между сервисами в сложных системах. Этим он нам и интересен.

Итак, HTTP — это протокол, который работает по принципу клиент-сервер. 



Это означает, что во взаимодействии участвует две программы, причём в разных ролях. Одна — клиент, или «заказчик услуг», формирует запрос и отправляет его к серверу. Сервер, или «поставщик услуг», получив запрос, обрабатывает его, формирует ответ и возвращает его клиенту. 

### А КАК КЛИЕНТ И СЕРВЕР ОБЩАЮТСЯ?

Запрос и ответ в HTTP являются строками, составленными в соответствии с протоколом из некоторых частей. Запрос состоит из трёх частей.



1) **Стартовая строка, или Request Line** — по ней определяется вид запроса.  
2) **Заголовки запроса, или Request Headers** — дополнительные параметры запроса, в которых обычно передаётся служебная информация. Например, в каком формате ожидается ответ, или информация о клиенте.   
3) **Тело запроса, или Request Message Body** — содержит данные для передачи. Эта часть присутствует не всегда.  

Давайте чуть подробнее разберём первую часть.   



Первое, что в ней указано — это **метод, или тип запроса**. Есть набор стандартных методов, но теоретически вы можете делать свои. 

#### Основные методы: 

GET — обычно означает получение содержимого ресурса и не содержит тела.   
POST — наоборот, передача данных ресурсу.  
PUT — обновление ресурса.  
DELETE — удаление ресурса.  



В HTTP мы работаем с ресурсами, которые расположены по некоторому адресу на сервере. Изначально под ресурсами понимались HTML-файлы на сайте, но сейчас это уже некоторое абстрактное понятие. 

Адрес ресурса, или _URI (Uniform Resource Identifier)_ — это то, что вы видите в адресной строке браузера. Он следует за методом в стартовой строке запроса. 



Последней идёт версия HTTP-протокола (кстати, последняя, 1.1, появилась в 1999 году).



Ответ также состоит из стартовой строки, заголовков и тела. 



Основное отличие — в стартовой строке: там вместо метода и URI указывается код состояния. Это численное значение, которое показывает результат обработки. Коды задаются протоколом, и вы наверняка встречали «ошибку 404» на сайтах. 


**404** — это как раз код состояния, означающий, что ресурса с заданным URI не существует на сервере. 

**Группы кодов состояния ответа HTTP-сервера делятся на следующие группы:**

- Информационные (100–199).
- Успешно (200–299).
- Перенаправление (300–399).
- Ошибка клиента (400–499).
- Ошибка сервера (500–599).

Например:

Код 200 означает, что запрос обработан успешно.

Код 500 означает, что сервер вообще не смог обработать запрос. 

### REST



Cпецификация HTTP не обязывает сервер понимать все методы, а также не указывает серверу, что он должен делать при получении запроса с тем или иным методом. Поэтому был изобретён архитектурный стиль **REST.**

Он даёт нам более верхнеуровневые указания, чем HTTP-протокол, а именно:

- как правильно организовывать адресацию к ресурсам;
- какие методы у этих ресурсов должны быть;
- какой ожидается результат.  

Основная концепция философии REST заключается в том, что клиентом RESTful-сервера может быть кто угодно: браузер, другое приложение, разработчик. Веб-приложение, спроектированное по правилам REST, предоставляет информацию о себе в форме информации о своих ресурсах. 

Ресурс может быть любым объектом, о котором сервер предоставляет информацию. Например, в API Instagram ресурсом может быть пользователь, фотография, хэштег. Каждая единица информации (ресурс) однозначно определяется URL.

- GET-запрос /rest/users — получение информации о всех пользователях
- GET-запрос /rest/users/125 — получение информации о пользователе с id=125
- POST-запрос /rest/users — добавление нового пользователя
- PUT-запрос /rest/users/125 — изменение информации о пользователе с id=125
- DELETE-запрос /rest/users/125 — удаление пользователя с id=125  

Поскольку спецификация является общепризнанной и широко распространённой, то следовать ей очень полезно. 

Если вы говорите, что ваш сервис — RESTful (то есть построен по этой спецификации), то любой разработчик сразу поймёт, как примерно устроен ваш сервис. А значит, он сможет быстро начать им пользоваться: отправлять правильные запросы и получать тот результат, который он ожидает. 

### ПОДВЕДЁМ ИТОГИ

Для того чтобы наладить взаимодействие по HTTP, на стороне клиента вам нужно сформировать запрос: указать адрес, куда отправится запрос, выбрать метод, а также задать заголовки и тело запроса, если это нужно для выбранного метода. 

На стороне сервера при получении запроса нужно знать, какой ресурс запрошен, чтобы понять, каким кодом его обрабатывать. А этот код, в зависимости от того, что он должен делать, может проверить значения заголовков и/или прочитать тело запроса.
