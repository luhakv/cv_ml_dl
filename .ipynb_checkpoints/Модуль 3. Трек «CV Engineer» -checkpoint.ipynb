### 2. О треке «CV Engineer»
→ Для начала предлагаем вам ознакомиться с программой трека. Обращаем внимание, что выделенные тематические блоки не равны по объёму и длительности — конечно же, основной фокус трека направлен на практическое применение знаний, и таким материалам отведена наибольшая часть времени обучения.

#### БЛОК 1. РАБОТА С ИЗОБРАЖЕНИЯМИ И МАТЕМАТИКА НЕЙРОСЕТЕЙ

- научитесь базовой работе с изображениями: предобработка, трансформация, использование сервисов для сбора и разметки;
- разберёте основные математические понятия в работе с нейросетями.

#### БЛОК 2. HARD & SOFTWARE

- научитесь оценивать вычислительные ресурсы под задачу;
- изучите работу с облачными серверами;
- познакомитесь с основными библиотеками для нейронных сетей, поймёте их основные особенности и разницу между ними;
- научитесь распараллеливать вычисления и разберётесь, в каких частях кода это возможно.

#### БЛОК 3. НЕЙРОННЫЕ СЕТИ ДЛЯ CV: БАЗОВЫЕ ЗАДАЧИ

- детально разберётесь в архитектурах свёрточных сетей и научитесь решать задачу классификации изображений;
- поймёте, как пользоваться предобученными сетями и дообучать сети под задачу;
- научитесь решать задачу распознавания объектов с использованием разных архитектур;
- научитесь решать задачу сегментации с использованием разных архитектур.

#### БЛОК 5. НЕЙРОННЫЕ СЕТИ В ПРОДАКШЕНЕ

- научитесь выводить модели в продакшен и мониторить их работу;
- погрузитесь в тонкости продакшена со стороны ресурсов и платформ.

#### БЛОК 6. REAL LIFE CV-ИНЖЕНЕРА

- рассмотрите, как выглядит работа инженера по компьютерному зрению в реальной жизни;
- поймёте подводные камни и боли работы с реальными датасетами и бизнес-задачами;
- разберёте реальные кейсы специалистов из области.
